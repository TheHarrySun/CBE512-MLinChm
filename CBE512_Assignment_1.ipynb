## **(b)** Processing strings into feature vectors.
Write a function that converts the SMILES string to a numerical "Bag-of-Elements" vector.  Note that you will have to be a little careful with handling 'Br' vs. 'B' and 'Cl' vs. 'C'b

```python
# MY CODE

# collect the data from the file
url_of_smiles = "https://raw.githubusercontent.com/webbtheosim/CBE512-MLinChmSciEng/main/data/Tm_200_subset.csv"
data = pd.read_csv(url_of_smiles, sep = ',')

# define a function that converts the SMILES to a vector with the number of each element
def convertSMILESToVector(smiles_string):
  ans = [0] * 10
  elements = ['C', 'O', 'N', 'l', 'B', 'r', 'I', 'F', 'S', 'P']
  for char in smiles_string:
    for i in range(0, len(elements)):
      if (elements[i] == char):
        ans[i] += 1
  ans[0] -= ans[3]
  ans[4] -= ans[5]
  return ans

# collect all of the data from the file and put it into a variable smiles_data
smiles_data = data["smiles"].to_numpy()
```

## **(c)** Chemical similarity as a norm.

Use the norm induced by the dot product to obtain a  distance as a metric for similarity / dissimilarity. Identify which set of molecules are "most similar" and which set of molecules are "most dissimilar" based on this distance metric. What limitations might there be to using this expression for similarity between molecules?

```python
# MY CODE

import math

# define a function that calculates the distance, which I put as dot product of the two vectors divided by the length of each vector
def calculate_distance(mol1, mol2):
  length1 = 0
  length2 = 0
  vector1 = convertSMILESToVector(mol1)
  vector2 = convertSMILESToVector(mol2)

  # calculate the length of each vector
  for i in range(0, len(vector1)):
    length1 += vector1[i] ** 2
    length2 += vector2[i] ** 2
  length1 = math.sqrt(length1)
  length2 = math.sqrt(length2)

  # calculate the dot product
  dot = 0
  for i in range(0, len(vector1)):
    dot += vector1[i] * vector2[i]

  # return the dot product divided by the lengths of both vectors
  dot /= length1 * length2
  return dot

# calculate the pairwise distances between all of the data in the smiles_data
distances_btwn_mols = []
for i in range(0, len(smiles_data)):
  for j in range(i, len(smiles_data)):
    distances_btwn_mols.append(calculate_distance(smiles_data[i], smiles_data[j]))
```

The molecules that have the closest molecular formulas are deemed the most similar and therefore get a distance close to 1, whereas the most disimilar molecules are the ones that are considered orthogonal so that their distance becomes close to 0.

The limitations for using this expression for similarity is that it only takes into account the number of atoms of each element and does not include the structural complexity. As a result, this model ends up simplifying the data and not taking into account all aspects of the molecules to determine molecular complexity.

Repeat parts **(c)** and **(d)** for a *different* distance metric of your choice. You might look into topics like Earth mover's distance or Tanimoto similarity/Jaccard index. How does this impact your findings? Your distance metric need not have an associated inner product.


```python
# MY CODE

# create a function that calculates the jaccard index
# my method for calculating jaccard index only takes into account the union and intersection of the atom numbers
# so it doesn't include similarity from molecular structure, only atoms in the molecule
def calculate_jaccardindex(mol1, mol2):
  vector1 = convertSMILESToVector(mol1)
  vector2 = convertSMILESToVector(mol2)
  union = 0
  intersection = 0
  for i in range(0, len(vector1)):
    if (vector1[i] > vector2[i]):
      union += vector1[i]
      intersection += vector2[i]
    else:
      union += vector2[i]
      intersection += vector1[i]
  return intersection / union

# calculate the pairwise jaccard indices from the data
jaccard_distances = []
for i in range(0, len(smiles_data)):
  for j in range(i, len(smiles_data)):
    jaccard_distances.append(calculate_jaccardindex(smiles_data[i], smiles_data[j]))

# plot the jaccard indices
plt.hist(jaccard_distances, bins = 250, range=[0, 1])

plt.xlabel('Jaccard Index Between Molecules')
plt.ylabel('Frequency')
plt.title('Frequency of Jaccard Indices Between Molecules')

plt.show()
```

$$\tilde{\text{MW}} = \frac{\text{MW} - \mu_\text{MW}}{\sigma_{\text{MW}}}$$

where $\mu_x$ and $\sigma_x$ indicate the mean and standard deviation of variable $x$.

Plot the distribution of $\tilde{\text{MW}}$. How does this compare to the original distribution?

```python
# define function that normalizes the data from a dictionary
def standard_normalization(data):
  data_values = [float(elem) for elem in list(data.values())]
  mean = sum(data_values) / len(data_values)

  variance = 0
  for i in data_values:
    variance += (i - mean) ** 2
  variance /= len(data_values)

  stddev = math.sqrt(variance)

  normalized = [(elem - mean) / stddev for elem in data_values]
  return normalized

# normalize the two dictionaries
normalized_smiles = standard_normalization(smiles_mw_dict)
normalized_carbonyls = standard_normalization(carbonyl_dict)

# plot the histograms
plt.hist(normalized_smiles, bins = 250, alpha = 0.45, color = 'r')
plt.hist(normalized_carbonyls, bins = 250, alpha = 0.45, color = 'b')

plt.xlabel('Molecular Weight')
plt.ylabel('Frequency')
plt.legend(['Normalized Full Dataset', 'Normalized Carbonyl Dataset'])
plt.title('Normalized Frequency of Molecular Weights in Full Dataset and Carbonyl Dataset')

plt.show()
```

Consider the two functions below that will effectively do the same thing: take as input a dictionary and output a dictionary with all the values doubled.

```python
def double_it(the_dict):
  new_dict = {}
  for key,val in the_dict.items():
    new_dict[key] = 2*val
  return new_dict

def double_in_place(the_dict):
  for key in the_dict:
    the_dict[key] *= 2
  return
```

```python
import random
import numpy as np
myDict  = {i: np.round(random.random(),3) for i in range(10)}
print(myDict)

doubleDict = double_it(myDict) # here we create a new dictionary and send it out
print(doubleDict)

double_in_place(myDict) # here we do not have an output, we have modified the object coming in
print(myDict)

alsoDoubled = {key: 2*val for (key,val) in myDict.items()} # here we avoid creating a function
print(alsoDoubled)
```
